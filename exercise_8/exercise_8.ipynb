```python
import pandas as pd
import numpy as np

df = pd.read_csv("West_Nile_Virus_by_County.csv")
```

```python
counties = df["County"].unique()
```

```python
from tqdm import tqdm
location = {}
for idx in tqdm(range(counties.shape[0])):
    g = geocoder.osm('{} County, CA'.format(counties[idx]))
    location[counties[idx]] = (g.geojson['features'][0]['geometry']['coordinates'])
```

```python
countyDF = pd.DataFrame.from_dict(location, orient='index').reset_index()
countyDF.columns = ["County", "Latitude", "Longitude"]
```

```python
countyDF
```

```python
dfMerged = df.merge(countyDF)
```

```python
dfData = dfMerged[["Longitude", "Latitude"]].values
data = []
for idx in range(dfData.shape[0]):
    data.append(list(dfData[idx]))
```

```python
from folium.plugins import HeatMap, HeatMapWithTime

m = folium.Map([35., -118.5], tiles='stamentoner', zoom_start=5)

HeatMap(data).add_to(m)

#HeatMapWithTime(data, index=list(dfMerged["Year"].values)).add_to(m)
m.save('Heatmap.html')

m
```
